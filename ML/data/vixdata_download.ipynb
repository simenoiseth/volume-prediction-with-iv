```python
import pandas as pd
from pandas_datareader import data as pdr

start = "2004-01-01"
end   = None  # None = up to today

vix = pdr.DataReader("VIXCLS", "fred", start=start, end=end)
vix = vix.rename(columns={"VIXCLS":"vix_close"})

# save
vix.to_csv("vix_closing_prices_2004_to_today_FRED.csv", index_label="date")
print(vix.head(), vix.tail())
```
