```python
import wrds, pandas as pd
db = wrds.Connection()

# A) See all libraries you can access
libs = db.list_libraries()
print("Libraries you can see:", libs)

# B) List all schemas that look like TAQ; works even if some are blocked
taq_schemas = [lib for lib in libs if lib.lower().startswith('taq')]
print("Candidate TAQ schemas:", taq_schemas)

# C) Within each candidate schema, list some tables (skip on privilege errors)
for sch in taq_schemas:
    try:
        tbls = db.list_tables(sch)
        print(f"{sch}: {len(tbls)} tables (showing up to 20):", tbls[:20])
    except Exception as e:
        print(f"{sch}: cannot list tables ({e})")

# D) If your WRDS supports querying pg_catalog, this finds any CTM trades tables you can actually SELECT
try:
    probe = db.raw_sql("""
        SELECT schemaname, tablename
        FROM pg_tables
        WHERE schemaname ILIKE 'taq%%'
          AND tablename ILIKE 'ctm_%%'
        ORDER BY schemaname, tablename
        LIMIT 200
    """)
    print(probe)
except Exception as e:
    print("pg_catalog probe skipped:", e)
```
