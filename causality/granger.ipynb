```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
df = pd.read_csv("volume_vix_merged.csv", parse_dates=["date"])
data = df[["sh_volume", "vix_close"]].dropna()

max_lag = 5
test_result = grangercausalitytests(data, max_lag, verbose=True)
```

```python
for lag, res in test_result.items():
    f_test = res[0]['ssr_ftest']
    f_value, p_value = f_test[0], f_test[1]
    print(f"Lag {lag}: F = {f_value:.6f}, p = {p_value:.20f}")
```
